```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/1+np.exp(-x)
x=np.array([1,50,100,1000])
y=sigmoid(x)
plt.plot(x,y)

```

```python
def tanh(x):
  return np.tanh(x);
x=np.array([1,50,100,1000])
y=tanh(x)
plt.plot(x,y)
```

```python
def relu(x):
  return np.maximum(x,0)
x=np.array([100,50,100,1000])
y=relu(x)
plt.plot(x,y)
```

```python
def softmax(x):
  p=np.exp(x-np.max(x))
  return p/np.sum(p)
x=np.array([100,50,100,1000])
y=softmax(x)
plt.plot(x,y)
```
